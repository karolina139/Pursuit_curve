# Krzywe pościgowe

### WSTĘP
Problem krzywych zakreślonych podczas pościgu interesował człowieka już wiele lat temu. Z krzywymi pościgowymi mamy do czynienia na co dzień. Jako przykład możemy podać pogoń psa za zającem, czy też znaną nam z młodszych lat zabawę w berka.

Matematyczne roważania dotyczące pościgu po raz pierwszy zostały podjęte przez Pierre'a Bouguera w XVIII wieku. Analizował on pościg okrętu pirackiego za statkiem handlowym, a jego rozważania opublikowano w 1732 r. [1] Na kartach historii zapisał się także George Boole, który zdefiniował pojęcie krzywej pościgowej w swoim dziele "Treatise on differential equations" (1859 r.)[2]. Podjął w nim zagadnienie samolotu przeprowadzajacego atak na ruchomy cel przy zachowanu odpowiedniego limitu celowania. Tym samym dał nam przykład zastosowania krzywych pościgowych w obszarze działań wojskowych. Okazuje się, że krzywe pościgowe są wykorzystywane także przez NASA, na przykład podczas przewidywania zderzeń ciał niebieskich. Jak możemy zauważyć, temat krzywych pościgowych jest bardzo istotny, a jego zastosowania odnajdziemy w wielu dziedzinach.

Podstwową definicję krzywej pościgowej możemy zapisać następująco

> #### Def.  Krzywa pościgowa
>Krzywa pościgowa to krzywa zakreślona przez ciało A podczas pogoni za ciałem B.

Zajmiemy się przede wszystkim ruchem w dwóch wymiarach, natomiast analizę krzywych pościgowych można rozszerzać do trzech wyniarów. Możemy także modyfikować liczbę obektów ścigających oraz zmieniać prawo, wedle którego się poruszają, na przykład poprzez opisanie ruchu wybraną funkcją matematyczną.

Celem projektu jest przedstawienie wybranych zagadnień związanych z krzywymi pościgowymi. Do analizy danych problemów będziemy wykorzystywać równania różniczkowe pierwszego  i drugiego rzędu raz układy równań różniczkowych. Aby je rozwiązać zastosujemy metody numeryczne takie jak metoda Eulera, metoda Adamsa – Bashfortha czy Adamsa-Moultona.

### METODY NUMERYCZNE

Niech równanie różniczkowe dane będzie w postaci $y'(t) = f(y, t)$, $y(0) = y_0$ a $h$ to krok całkowania. Wtedy rozwiązanie tego równania możemy wyznaczyć za pomocą następujących metod.

>#### Metoda Eulera
> $y_{n+1} = y_n + h \cdot f_n$

>#### Metoda Adamsa-Bashfortha drugiego rzędu
> $y_{n+1} = y_n + 0,5h \cdot(3f_n - f_{n-1})$


>#### Metoda Adamsa-Moultona drugiego rzędu
> $y_{n+1} = y_n + 0,5h \cdot(f_{n+1} + f_{n})$

### PROSTA KRZYWA POGONI

> #### Def. Prosta krzywa pogoni
> Prosta krzywa pogoni określona jest przez przypadek, gdy ciało uciekające B porusza się po linii prostej.

Pierre Bouguer, jako hydrograf, analizował problem ucieczki statku handlowego przed okrętem pirackim. Rozważmy zatem następującą sytuację [1, 3]. Niech statek handlowy (B) porusza się z prędkością $v_B$ o stałej wartości wzdłuż prostej równoległej do osi $Oy$. W ogólnym przypadku możemy przyjąć, że statek i okręt piracki to punkty materialne. W chwili początkowej statek (B) znajduje się w punkcie $(x_0,0)$ na osi $Ox$, $x_0>0$, a okręt piracki (A), znajdujący się w początku układu współrzędnych $(0, 0)$, zaczyna płynąć z prędkością $v_A$ o stałej wartości w kierunku statku (B). Przyjmijmy, że współrzędne okrętu pirackiego (A) w dowolnej chwili to $(x, y)$.

<img src = "zdj/klasyczna_rys1.png" width="600" align="center">

Aby znaleźć równanie różniczkowe opisujące daną sytuację rozważmy najpierw trójkąt $AA_{y}B$ (rys. 1), w którym kąt $\alpha$ to kąt przy wierzchołku $A$. Droga przebyta przez $B$ w czasie $t$ to oczywiście $v_Bt$, stąd mamy

\begin{equation}
|AA_y| = x_0 - x,
\end{equation}

\begin{equation}
|BA_y| = v_Bt - y.
\end{equation}

Zauważmy, że prosta przechodząca przez wierzchołki $A$ i $B$ jest styczna do toru ruchu okrętu pirackiego w punkcie $(x, y)$. Możemy zatem wyznaczyć wspólczynnik kierunkowy tej prostej

\begin{equation}
\tag{1}
y' = \tan \alpha = \frac{v_{B}t - y}{x_0 - x}.
\end{equation}

Niech $s$ oznacza długość drogi pokonanej przez okręt piracki w czasie t. Wtedy mamy

\begin{equation}
\tag{2}
s = v_{A}t = \int_{0}^{x} \sqrt{1 + \left(\frac{dy}{du}\right)^2} du.
\end{equation}

Wyznaczając czas z równań (1) i (2) otrzymujemy

\begin{equation}
\frac{1}{v_A} \int_{0}^{x} \sqrt{1 + \left(\frac{dy}{du}\right)^2} du = \frac{y}{v_B} - \frac{x - x_0}{v_B}y'.
\end{equation}

Różniczkując powyższe równanie ze względu na zmienną x mamy

\begin{equation}
\frac{1}{v_A} \sqrt{1 + (y'(x))^2} = \frac{y'(x)}{v_B} - \frac{x - x_0}{v_B}y''(x) - \frac{y'(x)}{v_B}.
\end{equation}

Po prostych przekształceniach otrzymamy nieliniowe równanie różniczkowe drugiego rzędu, które opisuje tor ruchu okrętu pirackiego (A)

\begin{equation}
\tag{3}
y'' + \frac{v_B}{v_A} \frac{1}{x - x_0}  \sqrt{1 + (y')^2}  = 0.
\end{equation}

Z założeń początkowych wiemy, że w chwili początkowej okręt znajdował się w punkcie $(0, 0)$, a statek handlowy znajdował się na osi $Ox$. Stąd otrzymujemy warunki początkowe $y(0) = 0$, $y'(0) = 0$.

Aby rozwiązać równanie (3) wprowadzimy zmienną $u(x) = y'(x)$, $u'(x) = y''(x)$. W ten sposób dostajemy równanie różniczkowe pierwszego rzędu o zmiennych rozdzielonych

\begin{equation}
u' + \frac{v_B}{v_A} \frac{1}{x - x_0}  \sqrt{1 + u^2}  = 0.
\end{equation}

Rozdzielając zmienne mamy

\begin{equation}
\int \frac{du}{\sqrt{1 + u^2}}= \frac{v_B}{v_A}  \int \frac{1}{x_0 - x}dx.
\end{equation}

Następnie całkujemy obustronnie powyższe równanie

\begin{equation}
\log \left(u + \sqrt{1 + u^2}\right) = -\frac{v_B}{v_A} \log (x_0 - x) + C.
\end{equation}

Teraz zauważmy, że $u(0) = y'(0) = 0$, czyli

\begin{equation}
C = \frac{v_B}{v_A} \log x_0
\end{equation}

i ostatecznie

\begin{equation}
\log \left(u + \sqrt{1 + u^2}\right) = -\frac{v_B}{v_A} \log (x_0 - x) + \frac{v_B}{v_A} \log x_0.
\end{equation}





Korzystając z własności działań na logarytmach mamy kolejno

\begin{equation}
\log \left(u + \sqrt{1 + u^2}\right) = -\frac{v_B}{v_A} \log \frac{x_0 - x}{x_0},
\end{equation}

\begin{equation}
\log \left(u + \sqrt{1 + u^2}\right) + \log \left(\frac{x_0 - x}{x_0}\right)^\frac{v_B}{v_A} = 0,
\end{equation}

\begin{equation}
\log \left(\left(u + \sqrt{1 + u^2}\right) \cdot \left(\frac{x_0 - x}{x_0}\right)^\frac{v_B}{v_A} \right) = 0,
\end{equation}

\begin{equation}
\left(u + \sqrt{1 + u^2}\right) \cdot \left(\frac{x_0 - x}{x_0}\right)^\frac{v_B}{v_A} = 1,
\end{equation}

\begin{equation}
u + \sqrt{1 + u^2}= \left(\frac{x_0 - x}{x_0}\right)^{-\frac{v_B}{v_A} },
\end{equation}

\begin{equation}
\sqrt{1 + u^2}= \left(\frac{x_0 - x}{x_0}\right)^{-\frac{v_B}{v_A}} - u.
\end{equation}

Powyższe równanie podnosimy obustronnie do kwadratu otrzymując

\begin{equation}
1 + u^2= \left(\frac{x_0 - x}{x_0}\right)^{-\frac{2v_B}{v_A}} - 2u\left(\frac{x_0 - x}{x_0}\right)^{-\frac{v_B}{v_A}} + u^2.
\end{equation}

W ten sposób otrzymaliśmy równanie opisujące $u(x)$

\begin{equation}
u = \frac{1}{2} \left(\frac{x_0 - x}{x_0}\right)^{-\frac{v_B}{v_A}} - \frac{1}{2} \left(\frac{x_0 - x}{x_0}\right)^{\frac{v_B}{v_A}}.
\end{equation}

Wracając do zmiennej $y$ mamy

\begin{equation}
\tag{4}
\frac{dy}{dx} = \frac{1}{2} \left(\frac{x_0 - x}{x_0}\right)^{-\frac{v_B}{v_A}} - \frac{1}{2} \left(\frac{x_0 - x}{x_0}\right)^{\frac{v_B}{v_A}}.
\end{equation}

Otrzymaliśmy równanie różniczkowe pierwszego rzędu, które możemy rozwiązać numerycznie, stosując np. metodę Eulera. Aby rozwiązać to równanie analitycznie wystarczy rozdzielić zmienne i zcałkować obustronnie, w efekcie czego dostaniemy


\begin{equation}
y(x) = \frac{1}{2} \int \left(\frac{x_0 - x}{x_0}\right)^{-\frac{v_B}{v_A}} dx - \frac{1}{2} \int \left(\frac{x_0 - x}{x_0}\right)^{\frac{v_B}{v_A}} dx,
\end{equation}

a stąd ostatecznie

\begin{equation}
y(x) = -\frac{1}{2} \frac{x_0}{1 -\frac{v_B}{v_A}}\left(\frac{x_0 - x}{x_0}\right)^{1-\frac{v_B}{v_A}} + \frac{1}{2} \frac{x_0}{1 +\frac{v_B}{v_A}}\left(\frac{x_0 - x}{x_0}\right)^{1+\frac{v_B}{v_A}} + D,
\end{equation}

\begin{equation}
y(x) = \frac{x_0 - x}{2} \left(\frac{\left(\frac{x_0 - x}{x_0}\right)^{\frac{v_B}{v_A}}}{1 +\frac{v_B}{v_A}} - \frac{\left(\frac{x_0 - x}{x_0}\right)^{-\frac{v_B}{v_A}}}{1 - \frac{v_B}{v_A}}\right) + D.
\end{equation}

Uwzględniając warunek początkowy $y(0) = 0$ mamy

\begin{equation}
0 = \frac{x_0}{2} \left(\frac{1}{1 +\frac{v_B}{v_A}} - \frac{1}{1 - \frac{v_B}{v_A}}\right) + D,
\end{equation}

\begin{equation}
D = \frac{x_0}{2} \left(\frac{1}{1 -\frac{v_B}{v_A}} - \frac{1}{1 + \frac{v_B}{v_A}}\right) = \frac{x_0 \frac{v_B}{v_A}}{1 -\left(\frac{v_B}{v_A}\right)^2},
\end{equation}

a stąd dostajemy równanie opisujące trajektorię okrętu pirackiego

\begin{equation}
\tag{5}
y(x) = \frac{x_0-x}{2}\left(\frac{\left(\frac{x_0 - x}{x_0}\right)^{\frac{v_B}{v_A}}}{1 +\frac{v_B}{v_A}} - \frac{\left(\frac{x_0 - x}{x_0}\right)^{-\frac{v_B}{v_A}}}{1 - \frac{v_B}{v_A}}\right) + \frac{x_0 \frac{v_B}{v_A}}{1 -\left(\frac{v_B}{v_A}\right)^2}.
\end{equation}

### Przykład
Przyjmijmy, że początkowa odległość okrętu od statku w chwili początkowej jest równa $x_0 = 6$. Niech okręt piracki płynie z prędkością $v_A = 2$, a statek handlowy z prędkością $v_B = 1$. W takim przypadku pościg mógłby wyglądać następująco

```python
# %load skrypty/klasyczna_krzywa.py
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import math

def get_dataA(times, vA, xsB, ysB, tick):
    """
    Funkcja obliczająca współrzędne (x, y) obiektu goniącego.

    :param times: lista czasów
    :param vA: prędkość obiektu
    :param xsB: lista współrzędnych x obiektu uciekającego
    :param ysB: lista współrzędnych y obiektu uciekającego
    :param tick: krok czasowy
    :return: xs, ys
    """
    xs, ys = [], []
    for time in times:
        if time == 0:
            xs.append(0)
            ys.append(0)
        else:
            xA = xs[-1]
            yA = ys[-1]

            i = 0
            while times[i] != time:
                i += 1

            xB = xsB[i-1]
            yB = ysB[i-1]

            c = math.sqrt((yB-yA)**2 + (xB-xA)**2)

            xs.append(xA + (xB - xA)*vA*tick/c)
            ys.append(yA + (yB - yA)*vA*tick/c)

    return xs, ys


def get_dataB_classic(times, x0, vB):
    """
    Funkcja obliczająca współrzędne (x, y) obiektu uciekającego po linii prostej.

    :param times: lista czasów
    :param x0: odległość obiektu od osi Oy
    :param vB: prędkość obiektu
    :return: xs, ys
    """
    xs, ys = [], []
    for time in times:
        xs.append(x0)
        ys.append(vB*time)

    return xs, ys


def classic_curve(x0, vA, vB):
    """
    Funkcja tworząca animację pościgu.

    :param x0: położenie środka okręgu
    :param vA: prędkość obiektu goniącego
    :param vB: prędkość obiektu uciekającego
    :return: animacja
    """

    fig, ax = plt.subplots()
    xdataB, xdataA, ydataB, ydataA, xdataB2, xdataA2, ydataB2, ydataA2 = [], [], [], [], [0], [0], [0], [0]
    lnB, = plt.plot([], [], 'r', animated=True)
    lnB2, = plt.plot([], [], 'r', marker='o', animated = True)
    lnA, = plt.plot([], [], 'b', animated=True)
    lnA2, = plt.plot([], [], 'b', marker='o', animated=True)

    MAX_TIME = x0/(vA*(1 - (vB/vA)**2))
    NUMBER_OF_TIMES = 200

    times = np.linspace(0, MAX_TIME, NUMBER_OF_TIMES)

    xsB, ysB = get_dataB_classic(times, x0, vB)
    xsA, ysA = get_dataA(times, vA, xsB, ysB, MAX_TIME/NUMBER_OF_TIMES)

    def init():
        ax.set_xlim(0, x0 + 1)
        ax.set_ylim(0, vB*MAX_TIME + 1)
        ax.set_xlabel("x")
        ax.set_ylabel("y")
        ax.set_title("Klasyczna krzywa pogoni")
        lnB.set_data(xdataB, ydataB)
        lnA.set_data(xdataA, ydataA)
        lnB2.set_data(xdataB2, ydataB2)
        lnA2.set_data(xdataA2, ydataA2)
        return lnB, lnA, lnB2, lnA2


    def update(frame):
        i = 0
        while times[i] != frame:
            i += 1

        xdataB.append(xsB[i])
        xdataA.append(xsA[i])
        ydataB.append(ysB[i])
        ydataA.append(ysA[i])
        xdataB2[0] = xsB[i]
        xdataA2[0] = xsA[i]
        ydataB2[0] = ysB[i]
        ydataA2[0] = ysA[i]
        lnB.set_data(xdataB, ydataB)
        lnA.set_data(xdataA, ydataA)
        lnB2.set_data(xdataB2, ydataB2)
        lnA2.set_data(xdataA2, ydataA2)
        return lnB, lnA, lnB2, lnA2,

    ani = FuncAnimation(fig, update, frames=times,
                        init_func=init, blit=True, interval=2.5, repeat=False)

    # writer = PillowWriter(fps=25)
    # ani.save("classic_pursuit.gif", writer=writer)

    plt.show()


if __name__ == '__main__':
    classic_curve(6, 2, 1)
```

### KRZYWA POŚCIGU DLA ZADANEJ TRAJEKTORII

> ### Def. Krzywa pościgu dla zadanej trajektorii
> Krzywa pogoni określona dla przypadku, gdy trajektoria ciała uciekającego B jest zadana krzywą, a ciało ścigające A porusza się zawsze w stronę aktualnego położenia ciała B.

Tego typu rozważania będą nas prowadzić do analizy układu dwóch, na ogół nieliniowych rownań różniczkowych pierwszego rzędu, możliwych do rozwiązania jedynie metodami numerycznymi. [3]

<img src="zdj/ogolna_krzywa_pogoni1.png" width="600" align="center">

Rozważmy teraz sytuację ucieczki zająca przed psem. Niech zając (B) porusza się według z góry zadanej krzywej ze stałą prędkością $v_B$. Natomiast niech pies (A) startuje w chwili t = 0 ze znanego nam położenia $(x_a(t), y_a(t))$ i porusza sie ze stałą predkością $v_A$. Załóżmy że  $|v_A|=k|v_B|$ gdzie $k \in R^{+}$[1]. Położenia psa oraz zająca w układzie kartzjańskim zapiszmy używając wektorów jednostkowych

\begin{equation}
\overrightarrow{A(t)} = x_a(t) \hat i + y_a(t) \hat j,
\end{equation} 

\begin{equation}
\overrightarrow{B(t)} = x_b(t) \hat i + y_b(t) \hat j.
\end{equation}

Zapiszmy wektor jednostkowy $\overrightarrow{AB}$ (od psa w kierunku zająca)

\begin{equation}
\hat {AB} = \frac{\overrightarrow{B(t)}}{\overrightarrow{A(t)}} = \frac{\overrightarrow{OB} - \overrightarrow{OA}}{|\overrightarrow{OB} - \overrightarrow{OA}|}.
\end{equation} 

Jako że $|\frac{d\overrightarrow{A(t)}}{dt} |$jest szybkością psa, prędkość zapiszemy następująco

\begin{equation}
\frac{d\overrightarrow{A(t)}}{dt} = |\frac{d\overrightarrow{A(t)}}{dt}| \hat {AB}.
\end{equation}


Wiemy, że pies porusza sie k razy szybciej $|v_A|=k|v_B|$. Zatem

\begin{equation}
\frac{d\overrightarrow{A(t)}}{dt} = k|v_B | \hat {AB},
\end{equation}

\begin{equation}
\tag{6}
|v_B| = \sqrt{ \left(\frac{d x_B(t)}{dt}\right)^2  + \left(\frac{dy_B(t)}{dt}\right)^2 },
\end{equation}

\begin{equation}
\frac{d\overrightarrow{A(t)}}{dt} = k \sqrt{ \left(\frac{d x_B(t)}{dt}\right)^2  + \left(\frac{dy_B(t)}{dt}\right)^2 } \frac{\overrightarrow{OB} - \overrightarrow{OA}}{|\overrightarrow{OB} - \overrightarrow{OA}|}.
\end{equation}


Zapiszmy najpierw to równanie we współrzędnych kartzjańskich
\begin{equation}
\frac{d x_A(t)}{dt} \hat{i} + \frac{d y_A(t)}{dt} \hat{j}  = k \sqrt{ \left(\frac{d x_B(t)}{dt}\right)^2  + \left(\frac{dy_B(t)}{dt}\right)^2 } \frac{(x_B(t) - x_A(t))\hat{i}+ (y_B(t) - y_A(t))\hat{j}}{\sqrt{{(x_B(t) - x_A(t))}^2 + {(y_B(t) - y_A(t))}^2} }.
\end{equation}

Z powyższego równania otrzymujemy układ dwóch równań różniczkowych pierwszego rzędu, każde dla kolejnej składowej wektora prędkości $v_A$

\begin{equation}
\tag{7}
\dot x_A(t)  = k\sqrt{ {(\dot x_B(t))}^2 + {(\dot y_B(t))}^2 } \frac{x_B(t) - x_A(t)}{\sqrt{{(x_B(t) - x_A(t))}^2 + {(y_B(t) - y_A(t))}^2} },
\end{equation}

\begin{equation}
\tag{8}
\dot y_A(t)  = k\sqrt{ {(\dot x_B(t))}^2 + {(\dot y_B(t))}^2 } \frac{y_B(t) - y_A(t)}{\sqrt{{(x_B(t) - x_A(t))}^2 + {(y_B(t) - y_A(t))}^2}}.
\end{equation}

Powyższe równania rozwiązujemy numerycznie, podając tajektorię obiektu B.

Jednak te równania są zbyt ogólne, nie mówią nic o charakterze krzywej po której porusza się zając, ani o punktach startowych obu obiektów. Dalsze rozważania dla krzywej pościgu dla zadanej trajktorii przeprowadzimy dla ściśle zadanych krzywych obiektu uciekającego.

### Krzywa pościgu zadana trajektorią koła

Rozważmy tzw. "problem Hathway'a", w którym jeden z obiektów porusza się po okręgu (przeciwnie do wskazówek zegara), natomiast drugi - ścigający obiekt startuje ze środka koła i porusza się zawsze w kierunku pierwszego obiektu oraz niech prędkość drugiego obiektu do prędkości pierwszego będzie jak k:1. Wyznaczmy równania opisujące ruch obiektów, a także dystans jaki przebędzie drugi obiekt. [3]

Zapiszmy wcześniej wyprowadzone ogólne równania (7), (8) dla krzywej pościgu dla zadanej trajktorii. Niech obiektem B będzie kaczka, która porusza się wzdłuż brzegu jeziora, a obiektem A będzie pies. 

\begin{equation}
\dot x_A(t)  = k\sqrt{ {(\dot x_B(t))}^2 + {(\dot y_B(t))}^2 } \frac{x_B(t) - x_A(t)}{\sqrt{{(x_B(t) - x_A(t))}^2 + {(y_B(t) - y_A(t))}^2} },
\end{equation}

\begin{equation}
\dot y_A(t)  = k\sqrt{ {(\dot x_B(t))}^2 + {(\dot y_B(t))}^2 } \frac{y_B(t) - y_A(t)}{\sqrt{{(x_B(t) - x_A(t))}^2 + {(y_B(t) - y_A(t))}^2}}.
\end{equation}

Z treści problemu "Hatwaye'a" wiemy, że kaczka będzie się poruszać po okręgu. Załóżmy również, że kaczka startuje z punktu (1,0). Jej wektor zapiszemy następująco
\begin{equation}
\overrightarrow{B(t)} =cos(t) \hat i +sin(t) \hat j,
\end{equation} 

oraz (korzystając z (6))

\begin{equation}
|v_B| = \sqrt{ \left(\frac{d x_B(t)}{dt}\right)^2  + \left(\frac{dy_B(t)}{dt}\right)^2 } = \sqrt{ sin^2(t)  + cos^2(t) }  =1.
\end{equation}

Zatem bardziej szczegółowe rozwiązanie będzie miało postać

\begin{equation}
\dot x_A(t)  = k \frac{cos(t) - x_A(t)}{\sqrt{{(cos(t) - x_A(t))}^2 + {(sin(t) - y_A(t))}^2} },
\end{equation}

\begin{equation}
\dot y_A(t)  = k \frac{sin(t) - y_A(t)}{\sqrt{{(cos(t) - x_A(t))}^2 + {(sin(t) - y_A(t))}^2}}.
\end{equation}

Powyższy układ równań rozwiążemy numerycznie metodą ekstrapolacyjną Eulera oraz metodą Adamsa-Bashfortha.

Jeśli $n$ zdefiniujemy jako ilość boków to kąt ACB będzie zawsze równy $\frac{2 \pi}{n}$, gdzie $C$ to środek wielokąta foremnego a $A$ i $B$ dowolne dwa wierzchołki sąsiadujące ze sobą. Wzór na długość boku $a$ będzie następujący
<br><br>
\begin{equation}
\frac{\frac{a(0)}{2}}{r(0)} = sin(\frac{\pi}{n})  \rightarrow a(0) =  2r(0)sin(\frac{\pi}{n}).
\end{equation}
Zapiszmy również wzór na $v_r$(korzystamy z podobieństwa trójkątów)
\begin{equation}
v_r = \frac{dr}{dt} = -vsin(\frac{\pi}{n}).
\end{equation}
Minus wynika z tego, że wektor $v_r$ jest skierowany w stronę C, zatem r zmniejsza sie wraz z rosnącym czasem.
Obliczmy czas, po którym spotkają się myszy

\begin{equation} 
T = \int_{O}^{T}dt = \int_{r(0)}^{0}\frac{dr}{\frac{dr}{dt}} = \int_{r(0)}^{0}\frac{dr}{-vsin(\frac{\pi}{n})} = \frac{1}{v}\int_{0}^{r(0)}\frac{dr}{sin(\frac{\pi}{n})} = \frac{r(0)}{vsin(\frac{\pi}{n})}.
\end{equation}
Wiemy że $S = Tv$, zatem
\begin{equation}
S = \frac{r(0)}{sin(\frac{\pi}{n})}.
\end{equation}
Ze wzoru na długość boku $a$ otrzymujemy
\begin{equation}
r(0) =  \frac{a(0)}{2sin(\frac{\pi}{n})}.
\end{equation}
Ostatecznie długość trajektorii każdej z myszy będzie wyrażona wzorem
\begin{equation}
S = \frac{a(0)}{2sin^2(\frac{\pi}{n})}.
\end{equation}

### PODSUMOWANIE


Problem krzywych pościgowych jest bardzo złożony, stąd w niniejszym projekcie przedstawiona została jednynie namiastka możliwości, które daje nam ich analiza. Umiejętność matematycznego opisania pościgu umożliwia nam udzielenie odpowiedzi na konkretne pytania. Tak na przykład wyprowadziliśmy dokładne informacje dotyczące toru ruchu okrętu pirackiego. Mogliśmy także wyznaczyć czas trwania morskiego pościgu lub obliczyć drogę pokonaną przez każdą z goniących się nawzajem myszy.  Krzywe pościgowe dają też szerokie pole do eksperymentowania. Dzięki nim możemy tworzyć również efektowne animacje i rysunki. 

Przedstawione powyżej rysunki i animacje pościgów zostały wykonane z wykorzystaniem języka Python, a kody źródłowe dostępne są w folderze "skrypty".
